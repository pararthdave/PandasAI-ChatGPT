```python
import pandas as pd
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI
```

```python
from getpass import getpass
secret = getpass('Enter the API key value: ')
```

```python
# Sample DataFrame
df = pd.DataFrame({
    "country": ["United States", "United Kingdom", "France", "Germany", "Italy", "Spain", "Canada", "Australia", "Japan", "China"],
    "gdp": [19294482071552, 2891615567872, 2411255037952, 3435817336832, 1745433788416, 1181205135360, 1607402389504, 1490967855104, 4380756541440, 14631844184064],
    "happiness_index": [6.94, 7.16, 6.66, 7.07, 6.38, 6.4, 7.23, 7.22, 5.87, 5.12]
})

from pandasai.llm.openai import OpenAI
llm = OpenAI(api_token=secret)
pandas_ai = PandasAI(llm, conversational=False)
print(pandas_ai.run(df, prompt='Which are the 5 happiest countries?'))
```

```python
pandas_ai.run(
    df,
    "Plot the histogram of countries showing for each the happiness index, using different colors for each bar",
)
```

```python
pandas_ai.run(
    df,
    "Plot piechart of countries showing for each the gdp, using different colors for each bar",
)
```
